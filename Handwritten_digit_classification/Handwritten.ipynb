## clasification of hand written digits using neural newtork

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train,Y_train),(xtest,Ytest) = keras.datasets.mnist.load_data()
```

```python
x_train[0]
Y_train[0]
```

```python
%matplotlib inline
```

```python
''' Now we will reshape the xtrain and xtest dataset into a list of 1d array 
    so that each element is a node '''
```

```python
x_train = x_train/255
xtest = xtest/255
```

```python
flatxtrain = x_train.reshape(len(x_train),28*28)
flatxtest = xtest.reshape(len(xtest),28*28)
```

```python
''' now creating the model '''
```

```python
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation = 'sigmoid')
])
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
model.fit(flatxtrain,Y_train,epochs = 5)
```

```python
model.evaluate(flatxtest,Ytest)
```

```python
predictions = model.predict(flatxtest)
actual = [np.argmax(i) for i in predictions]
```

```python
cm = tf.math.confusion_matrix(labels = Ytest,predictions = actual)
```

```python
''' now visualizing the accuracy of the model using heatmap '''
import seaborn as sn
plt.figure(figsize = (8,5))
sn.heatmap(cm,annot = True, fmt = 'd')
plt.xlabel('predicted value')
plt.ylabel('actual value')
```

## Now creating the same type of the model but this time we will have a hidden layer as well

```python
model2 = keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation = 'relu'),
    keras.layers.Dense(10,activation = 'sigmoid')
])
model2.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
model2.fit(flatxtrain,Y_train,epochs = 5)
```

```python
predictions2 = model2.predict(flatxtest)
actual2 = [np.argmax(i) for i in predictions2]
```

```python
cm2 = tf.math.confusion_matrix(labels = Ytest,predictions = actual2)
```

```python
''' now visualizing the accuracy of the model using heatmap '''
import seaborn as sn
plt.figure(figsize = (8,5))
sn.heatmap(cm2,annot = True, fmt = 'd')
plt.xlabel('predicted value')
plt.ylabel('actual value')
```
