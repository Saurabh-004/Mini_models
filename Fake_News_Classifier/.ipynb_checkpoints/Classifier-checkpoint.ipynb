```python
import pandas as pd
import spacy 
import numpy as np

nlp = spacy.load("en_core_web_lg")
```

```python
fake = pd.read_csv("Fake.csv") # dataset containing all the fake News
true = pd.read_csv("True.csv") # dataset containing True News
```

```python
fake['Fake'] = 1 # creating a new column which stores fake as 1 and True as 0
true['Fake'] = 0
```

```python
df = pd.concat([fake,true],axis=0)  # concating both dataset into one
```

```python
df.head()
```

```python
df = df.sample(frac=0.7) # I used only 70% of the training data as it was taking plenty of time 
df.shape
```

```python
df = df.drop_duplicates(subset='text').reset_index(drop=True) # removing duplicates
```

```python
df['text'].duplicated().sum()
```

```python
# This will take time
# took around 1 hour so its just a waste of time 

# df['Vector'] = df['text'].apply(lambda x: nlp(x).vector) # changing all the text into their vector

# can use this method to save some time, this will use more ram and divide the task into 4 process, so RAM and CPU usage will go high and 
# your fan can enter into jet-mode

with nlp.select_pipes(enable=["tok2vec"]):
    vectors = [
        doc.vector
        for doc in nlp.pipe(
            df['text'],
            batch_size=100,
            n_process=1 # increase the process to 4 if you have high-end pc
        )
    ]

```

```python
# now create our feature and target variable

target = df['Fake']

# split it into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(vectors, target, 
                                                    random_state = 18, 
                                                    test_size = 0.2,
                                                    stratify = df['Fake'])
```

```python
# Now convert the train and test features into 2d array
X_train_2d = np.stack(X_train)
X_test_2d = np.stack(X_test)
```

```python
# The model will not support negative values that's why we scale our data using minmaxscaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaled_X_train_embed = scaler.fit_transform(X_train_2d)
scaled_X_test_embed = scaler.transform(X_test_2d)
```

```python
# now we will train the model
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC

NBClassifier = MultinomialNB() # naive bayes classifier
LSVC = LinearSVC() # linear Support Vector Classifier

LSVC.fit(scaled_X_train_embed,y_train)
NBClassifier.fit(scaled_X_train_embed,y_train)
```

```python
y_pred1 = LSVC.predict(scaled_X_test_embed)
y_pred2 = NBClassifier.predict(scaled_X_test_embed)
```

```python
## showing the model accuracy using classification report 
## precision, recall, and f1-score

from sklearn.metrics import classification_report, confusion_matrix


print(f'Linear SVC report \n {classification_report(y_test,y_pred1)}')
print(f'Naive bayes report \n {classification_report(y_test,y_pred2)}')
```

```python
# now ploting the confusion matrix to display the result
import seaborn as sns

cm = confusion_matrix(y_test,y_pred1)
sns.heatmap(cm,
            annot=True,
           fmt='d',
           cmap='Blues',
            xticklabels = ['Fake','Real'],
            yticklabels = ['Fake','Real']
            
           )
```

```python
cm = confusion_matrix(y_test,y_pred2)
sns.heatmap(cm,
            annot=True,
           fmt='d',
           cmap='Blues',
            xticklabels = ['Fake','Real'],
            yticklabels = ['Fake','Real']
            
           )
```
